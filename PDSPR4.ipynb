**1 . Explain the role of NumPy in the Python scientific computing ecosystem.**

Role of NumPy in the Python Scientific Computing Ecosystem

NumPy (Numerical Python) is the core library for numerical and scientific computing in Python.

Key roles:
Provides a powerful N-dimensional array object (ndarray)
Acts as a foundation for other libraries like Pandas, Matplotlib, and Scikit-learn
Supports fast mathematical operations on large datasets
Helps in working with:
Linear algebra
Statistics
Random number generation

**2.Why is NumPy Faster than Pure Python?**

NumPy is faster than regular Python because:

 a. Uses compiled code (C language)
NumPy operations are written in C, not pure Python
This makes execution much faster

b. Vectorization
Works on entire arrays at once instead of loops
Example:

for i in range(len(a)):

    c[i] = a[i] + b[i]

c = a + b

c. Less memory usage
Arrays store data more efficiently than Python lists

d. Better CPU usage
Uses optimized algorithms and memory handling

```python
import numpy as np
```

```python
arr1 = np.array([5, 10, 15, 20])
arr2 = np.arange(2, 20, 3)
arr3 = np.linspace(1, 10, 5)
arr4 = np.zeros((2, 3))
arr5 = np.ones((3, 3))
arr6 = np.eye(3)

print("Array:", arr1)
print("Arange:", arr2)
print("Linspace:", arr3)
print("Zeros:\n", arr4)
print("Ones:\n", arr5)
print("Identity Matrix:\n", arr6)
```

```python
import numpy as np

a = np.array([2, 4, 6, 8])
b = np.array([1, 3, 5, 7])

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Power:", a ** 2)
```

```python
import numpy as np

x = np.array([[1, 2], [3, 4]])
y = np.array([[5, 6], [7, 8]])

print("Matrix Multiplication:\n", np.dot(x, y))
print("Transpose of x:\n", x.T)
```

```python
import numpy as np

a = np.array([10, 20, 30, 40])
b = np.array([15, 20, 25, 40])

print("Greater than:", a > b)
print("Equal:", a == b)
print("Logical AND:", np.logical_and(a > 15, b > 20))
print("Logical OR:", np.logical_or(a < 20, b > 30))
```

```python
import numpy as np

a = np.array([1, 2, 3, 4])

print("Square Root:", np.sqrt(a))
print("Sine Values:", np.sin(a))
print("Log Values:", np.log(a))
print("Exponential:", np.exp(a))
```

```python
import numpy as np

arr = np.array([[4, 8, 12], [16, 20, 24]])

print("Sum:", arr.sum())
print("Row-wise Sum:", arr.sum(axis=1))
print("Column-wise Sum:", arr.sum(axis=0))
print("Maximum:", arr.max())
print("Minimum:", arr.min())
print("Mean:", arr.mean())
```

```python
import numpy as np

data = np.array([12, 18, 25, 30, 35, 40])

print("Median:", np.median(data))
print("Standard Deviation:", np.std(data))
print("Variance:", np.var(data))
print("Argmax:", np.argmax(data))
print("Argmin:", np.argmin(data))
```

```python
import numpy as np

a = np.array([9, 3, 6, 1, 9, 3, 7])

print("Sorted Array:", np.sort(a))
print("Unique Elements:", np.unique(a))
```
