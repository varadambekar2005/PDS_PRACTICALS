## Theory
**1.What is the importance of data visualization in data analysis? How does visualization help in exploratory data analysis (EDA)?**

Data visualization is one of the most important parts of data analysis because it helps convert raw numerical data into visual forms such as graphs, charts, and plots. Visual representation makes the data easier to understand and interpret.

In Exploratory Data Analysis (EDA), visualization helps to:

*identify trends and patterns
*detect outliers and anomalies
*understand data distribution
*compare different variables
*observe relationships between features
*support decision-making

For example, a scatter plot can show correlation between two variables, while a histogram helps in understanding frequency distribution.

**2.What is the difference between Matplotlib and other visualization libraries?**

Matplotlib is a foundational Python library used for creating static, animated, and interactive visualizations. Other libraries such as Seaborn, Plotly, and Pandas plotting functions are built on top of Matplotlib or provide specialized capabilities.

Difference:

*Matplotlib:	Basic and highly customizable plotting library
*Seaborn:	More attractive statistical graphics, built on Matplotlib
*Plotly:	Interactive and web-based visualizations
*Pandas: Plotting	Quick plotting directly from DataFrames

Matplotlib provides more control, while Seaborn makes complex statistical plots easier and more visually appealing.

**3.What is the purpose of the pyplot module in Matplotlib?**

The pyplot module is a collection of functions in Matplotlib that makes plotting simple and similar to MATLAB commands.

It is commonly imported as:

```
import matplotlib.pyplot as plt
```

The purpose of pyplot is to:

*create figures and axes
*plot graphs
*label axes
*set titles
*display plots
*customize appearance

Common functions include:

-plt.plot()
-plt.scatter()
-plt.bar()
-plt.hist()
-plt.show()

**4.Explain the process of creating a basic line plot**

The steps to create a basic line plot are:

Import Matplotlib
Define x-axis and y-axis data
Use plt.plot()
Add labels and title
Display graph using plt.show()

Example:

import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]

plt.plot(x, y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Basic Line Plot")
plt.show()

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/content/train (1).csv")

print(df.head())
````

````python
import matplotlib.pyplot as plt

plt.scatter(df['Age'], df['Fare'])
plt.title("Scatter Plot: Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
````

````python
plt.plot(df['PassengerId'], df['Fare'])
plt.title("Line Plot: Passenger ID vs Fare")
plt.xlabel("Passenger ID")
plt.ylabel("Fare")
plt.show()
````

````python
survival_count = df['Survived'].value_counts()

plt.bar(survival_count.index, survival_count.values)
plt.title("Bar Plot: Survival Count")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()
````

````python
plt.boxplot(df['Fare'])
plt.title("Box Plot: Fare Distribution")
plt.ylabel("Fare")
plt.show()
````

````python
plt.hist(df['Age'].dropna(), bins=10)
plt.title("Histogram: Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
````

````python
import seaborn as sns

sns.pairplot(df[['Age', 'Fare', 'Pclass', 'Survived']].dropna())
plt.show()
````
