**1 . What is data wrangling? Explain the difference between data wrangling and data preprocessing.**

Data Wrangling
Data wrangling means cleaning, organizing, and transforming raw data into a useful format for analysis.

It includes tasks like:

-Removing missing values

-Correcting wrong data

-Changing data types

-Renaming columns

-Combining multiple datasets

-Formatting data properly

Difference between Data Wrangling and Data Preprocessing

Data Wrangling:

-Focuses on cleaning and organizing raw data

-Includes handling missing values, duplicates, formatting

Data Preprocessing:

-Focuses on preparing data for machine learning or analysis

-Includes normalization, scaling, encoding

```python
import pandas as pd

df = pd.read_csv("/content/train (1).csv")
df.head()
```

```python
age_column = df["Age"]
print(age_column.head())
```

```python
df["Source"] = "Titanic Dataset"
print(df[["Name", "Source"]].head())
```

```python
sorted_df = df.sort_values(by="Fare", ascending=False)
print(sorted_df[["Name", "Fare"]].head())
```

```python
df_renamed = df.rename(columns={
    "Pclass": "Passenger_Class",
    "SibSp": "Siblings_Spouses",
    "Parch": "Parents_Children"
})
print(df_renamed.columns.tolist())
```

```python
print(df.describe())
```

```python
grouped = df.groupby("Pclass")[["Fare", "Age"]].mean()
print(grouped)
```

```python
extra_info = pd.DataFrame({
    "PassengerId": [1, 2, 3, 4, 5],
    "Nationality": ["British", "American", "Irish", "French", "German"]
})

merged_df = pd.merge(df, extra_info, on="PassengerId", how="left")
print(merged_df[["PassengerId", "Name", "Nationality"]].head())
```

```python
df_dropped_cols = df.drop(columns=["Cabin", "Ticket", "Source"])
print("After dropping columns:", df_dropped_cols.shape)
```

```python
df_dropped_rows = df_dropped_cols.dropna(subset=["Age"])
print("After dropping rows with missing Age:", df_dropped_rows.shape)
```

```python
def age_group(age):
    if age < 18:
        return "Child"
    elif age < 60:
        return "Adult"
    else:
        return "Senior"

df["Age_Group"] = df["Age"].apply(age_group)
print(df[["Name", "Age", "Age_Group"]].head(10))
```
